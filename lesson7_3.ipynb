```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')

student1 = {'':78, '':80, '':92, '':75, '':85}
student2 = {'':90, '':75, '':98, '':65, '':75}
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title("")
axes.set_xlabel("")
axes.set_ylabel("")
axes.plot(subjects,score1,'ro-.',label="")
axes.plot(subjects,score2,'gx--',label="")
axes.legend()
axes.grid(axis='y')

# subjects
math_index = subjects.index('')

for idx, (s, v1, v2) in enumerate(zip(subjects, score1, score2)):
    if s == '' and v1 == 92:
        # 
        # arc390
        axes.annotate(f'{v1}', 
                      xy=(math_index, v1), 
                      xytext=(math_index+0.5, v1-5),  # 
                      arrowprops=dict(
                          facecolor='red',
                          shrink=0.05,
                          connectionstyle='arc3,rad=-0.3'  # 
                      ))
        axes.annotate(f'{v2}', xy=(s, v2+0.5))
    else:
        axes.annotate(f'{v1}', xy=(s, v1+0.5))
        axes.annotate(f'{v2}', xy=(s, v2+0.5))
plt.show()
```
